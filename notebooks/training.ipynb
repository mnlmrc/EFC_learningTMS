```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import os
import globals as gl
```

```python
sn = 101
path = os.path.join(gl.baseDir, gl.behavDir, f'p{sn}_training', 'single_trial.tsv')
df = pd.read_csv(path, sep='\t')
df_s = df[df.success==1]
df_s = df_s.groupby(['BN', 'Chord']).mean(numeric_only=True).reset_index()

fig, ax = plt.subplots()

sb.lineplot(ax=ax, data=df_s, x='BN', y='ET', hue='Chord')
```
